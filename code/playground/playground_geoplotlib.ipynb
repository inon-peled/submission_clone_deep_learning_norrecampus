```python
import pandas as pd
import geoplotlib
from geoplotlib.utils import BoundingBox
import numpy as np
```

```python
dfdata = pd.read_csv('C:\\users\\inonpe\\Desktop\\Police_Department_Incidents_-_Current_Year__2017_.csv', 
                     parse_dates=['Date'])
dfdata.head()
```

```python
geo_data_for_plotting = {'lat': dfdata['Y'], 'lon': dfdata['X']}
```

```python
#create the BoundingBox
max_lat = np.max(dfdata['Y'])
min_lat = np.min(dfdata['Y'])
max_lon = np.max(dfdata['X'])
min_lon = np.min(dfdata['X'])
print max_lat, min_lat, max_lon, min_lon
bbox = BoundingBox(north=max_lat, west=min_lon, south=min_lat, east=max_lon)
geoplotlib.set_bbox(bbox)
# geoplotlib.set_bbox(BoundingBox.KBH)
geoplotlib.kde(geo_data_for_plotting, bw=5, cmap='hot', cut_below=1e-4)

geoplotlib.inline()
geoplotlib.show()
```
